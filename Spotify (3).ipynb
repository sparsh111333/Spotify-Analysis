```python
def categorize_popularity(x):
    if x >= 80:
        return 'Very Popular'
    elif x >= 60:
        return 'Popular'
    elif x >= 40:
        return 'Average'
    else:
        return 'Low'

Spotify['popularity_category'] = Spotify['popularity'].apply(categorize_popularity)
```

```python
Spotify
```

```python
Spotify['danceability_cat'] = pd.cut(
    Spotify['danceability'],
    bins=[0, 0.4, 0.7, 1],
    labels=['Low', 'Medium', 'High']
)
```

```python
Spotify
```

```python
Spotify['energy_category'] = pd.cut(
    Spotify['energy'],
    bins=[0, 0.4, 0.7, 1],
    labels=['Calm', 'Moderate', 'High']
)
```

```python
Spotify['Valence_category'] = pd.cut(
    Spotify['valence'],
    bins=[0, 0.3, 0.6, 1],
    labels=['Sad', 'Neutral', 'Happy']
)
```

```python
Spotify['tempo_cat'] = pd.cut(
    Spotify['tempo'],
    bins=[0, 90, 120, 150,250],
    labels=['Slow', 'Medium', 'Fast',"very fast"]
)
```
